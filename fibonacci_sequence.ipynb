```python
def fibonacci(n):
    if n<=0:
        print("Please enter a positive integer.")
    elif n == 1:
        return[0, 1]
    else:
        sequence = fibonacci(n-1)
        sequence.append(sequence[-1] + sequence[-2])
        return sequence
        
n = int(input("Enter number of terms: "))
result = fibonacci(n)
print("Fibonacci sequence of", n,": ", result)
```
